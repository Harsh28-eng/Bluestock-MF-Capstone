```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go
```

```python
fund_master = pd.read_csv("Data/Raw/01_fund_master.csv")
nav = pd.read_csv("Data/Raw/02_nav_history.csv")
aum = pd.read_csv("Data/Raw/03_aum_by_fund_house.csv")
sip = pd.read_csv("Data/Raw/04_monthly_sip_inflows.csv")
category = pd.read_csv("Data/Raw/05_category_inflows.csv")
folio = pd.read_csv("Data/Raw/06_industry_folio_count.csv")
performance = pd.read_csv("Data/Raw/07_scheme_performance.csv")
transactions = pd.read_csv("Data/Raw/08_investor_transactions.csv")
holdings = pd.read_csv("Data/Raw/09_portfolio_holdings.csv")
benchmark = pd.read_csv("Data/Raw/10_benchmark_indices.csv")

print("All datasets loaded successfully")
```

```python
nav["date"] = pd.to_datetime(nav["date"])

fig = px.line(
    nav,
    x="date",
    y="nav",
    color="amfi_code",
    title="NAV Trend 2022-2026"
)

fig.show()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=aum,
    x="year",
    y="aum_crore",
    hue="fund_house"
)

plt.title("AUM Growth by Fund House")
plt.show()
```

```python
fig = px.line(
    sip,
    x="month",
    y="sip_inflow_crore",
    title="Monthly SIP Inflows"
)

fig.show()
```

```python
pivot = category.pivot_table(
    values="net_inflow",
    index="category",
    columns="month"
)

sns.heatmap(pivot)
plt.show()
```

```python
transactions["age_group"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.show()
```

```python
sns.boxplot(
    data=transactions,
    x="age_group",
    y="amount_inr"
)
plt.show()
```

```python
sns.boxplot(
    data=transactions,
    x="age_group",
    y="amount_inr"
)
plt.show()
```

```python
transactions.groupby(
    "state"
)["amount_inr"].sum().sort_values().plot(
    kind="barh"
)

plt.show()
```

```python
transactions["city_tier"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.show()
```

```python
plt.plot(
    folio["date"],
    folio["folio_count"]
)

plt.show()
```

```python
corr = performance.corr(
    numeric_only=True
)

sns.heatmap(
    corr,
    annot=True
)

plt.show()
```

```python
sector = holdings.groupby(
    "sector"
)["weight_pct"].sum()

plt.pie(
    sector,
    labels=sector.index
)

centre = plt.Circle(
    (0,0),
    0.70,
    fc="white"
)

plt.gca().add_artist(centre)

plt.show()
```

```python
top10 = performance.sort_values(
    "return_5yr_pct",
    ascending=False
).head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    data=top10,
    x="return_5yr_pct",
    y="scheme_name"
)

plt.title("Top 10 Funds by 5-Year Return")

plt.savefig("../Reports/13_top_10_returns.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=performance,
    x="risk_score",
    y="return_5yr_pct"
)

plt.title("Risk vs Return")

plt.savefig("../Reports/14_risk_return_scatter.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    nav["nav"],
    bins=30
)

plt.title("NAV Distribution")

plt.savefig("../Reports/15_nav_distribution.png")

plt.show()
```
